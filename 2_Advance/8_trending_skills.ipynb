```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda skill : ast.literal_eval(skill) if pd.notna(skill) else skill)
```

```python
df_exploded=df.explode('job_skills')
```

```python
df_exploded['month_no']=df_exploded['job_posted_date'].dt.month
df_exploded=df_exploded[df_exploded['job_title_short']=='Data Analyst']
```

```python
df_pivot=df_exploded.pivot_table(index='month_no',columns='job_skills',aggfunc='size',fill_value=0)
df_pivot
```

```python
df_pivot.loc['total']= df_pivot.sum()
df_pivot=df_pivot[df_pivot.loc['total'].sort_values(ascending=False).index]
df_pivot=df_pivot.drop('total')
df_pivot
```

```python
df_pivot=df_pivot.reset_index()
df_pivot['job_posted_month']=df_pivot['month_no'].apply(lambda x : pd.to_datetime(x,format='%m').strftime('%b'))
df_pivot=df_pivot.set_index('job_posted_month')
df_pivot=df_pivot.drop(columns='month_no')

df_pivot
```

```python
df_pivot.iloc[:,:5].plot(kind='line',
                         linewidth=4,
                         linestyle='-',# options: '-', '--', '-.', ':'
                         colormap='viridis', # options: 'viridis', 'plasma', 'inferno', 'magma', 'cividis'
                         markersize=6,
                         marker='o',# options: 'o', 'x', '+', '*', 's', ',', '.', '1', '2', '3', '4', ...
                         figsize=(10,5))
plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```
