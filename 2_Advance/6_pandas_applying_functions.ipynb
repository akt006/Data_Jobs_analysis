```python
import pandas as pd
from datasets import load_dataset
import ast
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
```

```python
# Calculate projected salaries next year, using an assumed rate of 3.0% for all roles.
df_salary=df[pd.notna(df['salary_year_avg'])].copy()
def inflation(salary):
    return 1.03*salary
df_salary['inflated_salary']=df_salary['salary_year_avg'].apply(inflation)
df_salary[['salary_year_avg','inflated_salary']]
```

```python
# Calculate projected salaries next year, but:

# For senior roles (e.g., Senior Data Analysts), assume the rate is 5%
# For all other roles, assume rate is 3%
def cat_inflation(row):
    if 'Senior' in row['job_title_short']:
        return 1.03*row['salary_year_avg']
    else:
        return 1.05*row['salary_year_avg']

df_salary['updated_inflation']=df_salary.apply(cat_inflation,axis=1)

df_salary[['job_title_short','salary_year_avg','updated_inflation']]
```

```python
# Convert the job_skills from a generic object to an actual list object (hint this is very important for later). Let's try doing that by just using ast.literal_eval and then look at our new column.
df_updated=df.copy()
def tolist(skills):
    if pd.notna(skills):
        return ast.literal_eval(skills)
df_updated['job_skills']=df_updated['job_skills'].apply(tolist)
df_updated['job_skills'][0]
```
