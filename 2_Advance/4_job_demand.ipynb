```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
```

```python
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_india=df[df['job_country']=='India'].copy()
df_india
```

```python
df_india['month']=df_india['job_posted_date'].dt.strftime('%B')
job_demand_pivot=df_india.pivot_table(index='month',columns='job_title_short',aggfunc='size')
job_demand_pivot
```

```python
job_demand_pivot.reset_index(inplace=True)
job_demand_pivot
```

```python
job_demand_pivot['month_no']=pd.to_datetime(job_demand_pivot['month'],format='%B').dt.month
job_demand_pivot
```

```python

job_demand_pivot.set_index('month',inplace=True)
job_demand_pivot.sort_values(by='month_no',inplace=True)
job_demand_pivot=job_demand_pivot.drop(columns='month_no')
job_demand_pivot.rename_axis('job_posted_month',inplace=True)



```

```python
job_demand_pivot.plot(kind='line')
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1)) 
```

```python
top_3=df_india['job_title_short'].value_counts().head(3).index

job_demand_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the India')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

```python
india_dev_pivot=pd.read_csv('http://lukeb.co/software_csv',index_col='job_posted_month')
india_dev_pivot
```

```python

job_demand_pivot.columns
```

```python
merged_pivot=job_demand_pivot.merge(india_dev_pivot,on='job_posted_month')
merged_pivot
```

```python
top_5=merged_pivot.sum().sort_values(ascending=False).head().index
top_5
```

```python
merged_pivot[top_5].plot(kind='line')
plt.legend(loc='upper right', bbox_to_anchor=(1.45, 1)) 
```
