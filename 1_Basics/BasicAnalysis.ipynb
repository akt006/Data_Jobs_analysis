```python
import pandas as pd
from datasets import load_dataset
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df.head()
```

```python
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_posted_date']=df['job_posted_date'].dt.date
```

```python
df.head()
```

```python
us_jobs=df[df['job_country']=='United States']
us_jobs.head()
```

```python
us_jobs.dropna(subset=['salary_year_avg'],inplace=True)
us_jobs.head()
```

```python
us_jobs.groupby(['job_title_short'])[['salary_year_avg']].mean().sort_values(by='salary_year_avg',ascending=False)
```

```python
from matplotlib import pyplot as plt
df['salary_year_avg'].plot(kind='hist', bins=20, title='salary_year_avg')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
us_jobs.groupby('job_title_short').size().sort_values(ascending=False).head()
```

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median','min','max','count']).sort_values(by='median',ascending=False)
```
